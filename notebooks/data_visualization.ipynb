```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
```

## Test50-95 results

```python
df = pd.read_csv('test.csv')
```

```python
df.head()
```

```python
acc_df = df[df['metric']=='acc']
acc_df.head()
```

```python
loss_df = df[df['metric']=='loss']
loss_df.head()
```

```python
sns.set_style(style='white')
```

```python
p = sns.pointplot(y='value', x='prob', hue='method', data=acc_df, palette='mako')
p.set_xlabel("P(water background | water bird) = P(land background | land bird) at test time")
p.set_ylabel("Test Accuracy")
```

```python
p = sns.pointplot(y='value', x='prob', hue='method', data=loss_df, palette='mako')
p.set_xlabel("P(water background | water bird) = P(land background | land bird) at test time")
p.set_ylabel("Test Loss")
```

## Val

```python
df = pd.read_csv('val.csv')
```

```python
df.head()
```

```python
acc_df = df[df['metric']=='acc']
acc_df.head()
```

```python
loss_df = df[df['metric']=='loss']
loss_df.head()
```

```python
fig, ax = plt.subplots()
p = sns.pointplot(y='value', x='epoch', hue='method', data=acc_df, palette='mako')
p.set_xlabel("Epoch")
p.set_ylabel("Validation Accuracy")
ax.set_ylim(0.87, 0.97)
```

```python
fig, ax = plt.subplots()
p = sns.pointplot(y='value', x='epoch', hue='method', data=loss_df, palette='mako')
p.set_xlabel("Epoch")
p.set_ylabel("Validation Loss")
ax.set_ylim(0, 0.25)
```
